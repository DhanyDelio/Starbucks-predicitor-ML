```python
import pandas as pd 

X_train = pd.read_csv("Data/Processed_data/X_train.csv")
X_test = pd.read_csv("Data/Processed_data/X_test.csv")

y_train = pd.read_csv("Data/Processed_data/y_train.csv")
y_test = pd.read_csv("Data/Processed_data/y_test.csv")

print("Data has loaded")

```

```python
from sklearn.linear_model import LinearRegression

LR_model = LinearRegression()

LR_model.fit(X_train, y_train)

print("I am LR_model Know it")
```

```python
from sklearn import linear_model
lars = linear_model.Lars(n_nonzero_coefs=1).fit(X_train, y_train)

print("I am Lars know it")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(X_train,y_train)

print("I am GBR know it")
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

result_LR = LR_model.predict(X_test)
result_Lars = lars.predict(X_test)
result_GBR = GBR.predict(X_test)

data_eval = {
    'Model' : ['LinearRegression', 'lars', 'Gradiend Boosting'],
    'MAE' : [
        mean_absolute_error(y_test, result_LR),
        mean_absolute_error(y_test, result_Lars),
        mean_absolute_error(y_test, result_GBR)
    ],
    'R2 Scroe (%)' : [
        r2_score(y_test, result_LR)*100,
        r2_score(y_test, result_Lars)*100,
        r2_score(y_test, result_GBR)*100
    ]
}
df_result = pd.DataFrame(data_eval)
print("Comparison all model")
print(df_result)
```
